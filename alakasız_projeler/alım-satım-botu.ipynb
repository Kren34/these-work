```python
client = Client(API_KEY, API_SECRET)

# TRY fiyatÄ± iÃ§in USDTTRY dÃ¶viz kuru alÄ±nÄ±r
def get_usdt_try_price():
    ticker = client.get_symbol_ticker(symbol='USDTTRY')
    return float(ticker['price'])

# Coin'in USDT fiyatÄ±nÄ± al
def get_coin_price_in_usdt(symbol):
    try:
        ticker = client.get_symbol_ticker(symbol=f'{symbol}USDT')
        return float(ticker['price'])
    except:
        return 0.0

# TÃ¼m coinleri oku, USDT karÅŸÄ±lÄ±klarÄ±nÄ± topla
def get_wallet_value_in_try():
    balances = client.get_account()['balances']
    total_usdt_value = 0

    for asset in balances:
        coin = asset['asset']
        free = float(asset['free'])
        locked = float(asset['locked'])
        total = free + locked

        if total == 0 or coin == 'USDT':
            continue

        price = get_coin_price_in_usdt(coin)
        value_in_usdt = total * price
        total_usdt_value += value_in_usdt

    # EÄŸer USDT varsa onu da ekle
    for asset in balances:
        if asset['asset'] == 'USDT':
            total_usdt_value += float(asset['free']) + float(asset['locked'])

    # USDT â†’ TRY Ã§evir
    usdt_try = get_usdt_try_price()
    total_try = total_usdt_value * usdt_try
    return total_try

# SonuÃ§
wallet_try_value = get_wallet_value_in_try()
print(f"ðŸ’° CÃ¼zdan Toplam DeÄŸeri: {wallet_try_value:,.2f} TRY")
```

```python
client = Client(API_KEY, API_SECRET)

def get_usdt_try_price():
    try:
        ticker = client.get_symbol_ticker(symbol='USDTTRY')
        return float(ticker['price'])
    except:
        return 0.0

def get_coin_price_in_usdt(symbol):
    try:
        ticker = client.get_symbol_ticker(symbol=f'{symbol}USDT')
        return float(ticker['price'])
    except:
        return 0.0

def get_wallet_details_in_try():
    balances = client.get_account()['balances']
    usdt_try = get_usdt_try_price()

    data = []
    for asset in balances:
        coin = asset['asset']
        free = float(asset['free'])
        locked = float(asset['locked'])
        total = free + locked

        # Coin deÄŸeri sÄ±fÄ±rsa atla
        if total == 0:
            continue

        if coin == 'USDT':
            price_usdt = 1
        else:
            price_usdt = get_coin_price_in_usdt(coin)

        total_usdt = total * price_usdt
        total_try = total_usdt * usdt_try

        data.append({
            'Coin': coin,
            'Miktar': total,
            'Fiyat (USDT)': price_usdt,
            'Toplam (USDT)': total_usdt,
            'Toplam (TRY)': total_try
        })

    df = pd.DataFrame(data)
    df = df.sort_values(by='Toplam (TRY)', ascending=False)
    return df

# Ã‡alÄ±ÅŸtÄ±r
wallet_df = get_wallet_details_in_try()
print(wallet_df.to_string(index=False))
```
