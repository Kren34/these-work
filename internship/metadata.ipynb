```python
def find_folders_with_multiple_files(root_dir):
    folders_with_multiple_files = []

    for folder_name in os.listdir(root_dir):
        folder_path = os.path.join(root_dir, folder_name)
        if os.path.isdir(folder_path):  # Sadece klasörleri kontrol et
            file_count = len([f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))])
            if file_count > 1:
                folders_with_multiple_files.append(folder_name)

    return folders_with_multiple_files

# Ana dizin
data_dir = r"C:\Users\Yusuf\Desktop\staj\marmara\datalar\luad-lusc-karisik-islenmemis"

# Birden fazla dosya içeren klasörleri bul
folders = find_folders_with_multiple_files(data_dir)

print(folders)
```

```python
df = pd.DataFrame({'Klasör Adı': folders})

dosya_adlari = []
for klasor_adi in df['Klasör Adı']:
    klasor_yolu = os.path.join(data_dir, klasor_adi)
    tsv_dosyalari = [f for f in os.listdir(klasor_yolu) if f.endswith('.tsv')]
    if tsv_dosyalari:  # Klasörde en az bir .tsv dosyası varsa
        dosya_adlari.append(tsv_dosyalari[0])  # İlk .tsv dosyasının adını al
    else:
        dosya_adlari.append(None)  # .tsv dosyası yoksa None ekle

# Yeni sütunu DataFrame'e ekle
df['Dosya Adı'] = dosya_adlari

df1 = pd.DataFrame({'Klasör Adı': folders, 'Dosya Adı': dosya_adlari})
df1
```

```python
# normalde else kullanmak daha mantıklı olurdu ancak burada 3 farklı indirme yaptım, ve 3 farklı indirmeyi birbirleriyle kıyaslıyorum
# 1 tanesi luad-lusc karışık indirme, diğer ikisi teker reker indirme. eğer else kullansaydım istenmeyen bir verinin eklenmesi durumu
# yaşanma olasılığını tahmin ederek teker teker yaptım, ya da sadece elif de olabilirdi tabi ancak kopyalamak daha kolay geldi.

luadisimlerlist = luadisimler["Eski Dosya Adı"].tolist()
df1list = df1["Dosya Adı"].tolist()
ortaklarluad = []

for name in luadisimlerlist:
    if name in df1list:
        ortaklarluad.append(name)

luscisimlerlist = luscisimler["Eski Dosya Adı"].tolist()
df1list2 = df1["Dosya Adı"].tolist()
ortaklarlusc = []

for name in luscisimlerlist:
    if name in df1list2:
        ortaklarlusc.append(name)

len(ortaklarlusc), len(ortaklarluad)
```

```python
# "Kanser Tipi" sütununu oluştur ve varsayılan olarak "Bilinmiyor" ile doldur
df1["Kanser Tipi"] = "NaN"

# Dosya adına göre "Kanser Tipi" sütununu güncelle
for i, dosya_adi in enumerate(df1['Dosya Adı']):
    if dosya_adi in ortaklarlusc:
        df1.at[i, 'Kanser Tipi'] = "TCGA-LUSC"
    elif dosya_adi in ortaklarluad:
        df1.at[i, 'Kanser Tipi'] = "TCGA-LUAD"

df1
```

```python
# Dosyaların bulunduğu ana dizin
data_dir = r"C:\Users\Yusuf\Desktop\staj\marmara\datalar\luad-lusc-karisik-islenmemis"

# Tüm alt dizinleri ve dosyaları bul
all_dfs = []
for subdir, dirs, files in os.walk(data_dir):
    for file in files:
        if file == "annotations.txt":
            filepath = os.path.join(subdir, file)
            df = pd.read_csv(filepath, sep='\t')
            
            # Klasör adını ekle
            df["Klasör Adı"] = os.path.basename(subdir)  
            
            all_dfs.append(df)

# DataFrame'leri birleştir
combined_df = pd.concat(all_dfs, ignore_index=True)

# Sonuçları kaydet
combined_df.to_csv("combined_annotations.csv", index=False)
```
